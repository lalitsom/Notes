Statistics Definition
---------------------
Statistics is the study of how to collect and study data, so we can represent it in a simpler mathematical way.
Answer questions like, what is the average age of people in a city, or how many people have age 27, make predictions about future data.

First principles of statistics:
   - Represent data in a simpler way.
   - Extract parameters which best describe the data, and can be used to reconstruct it.
   - Make predictions about future data/whole population, based on past data/small sample.


Definitions of major terms
--------------------------
1. **Population**: A population is the entire group that you want to draw conclusions about. Represented by N.
2. **Sample**: A sample is the specific group that you will collect data from. Represented by n.
3. **Parameter**: A parameter is a number that describes the population. It is a fixed number, but in practice we do not know its value. example: average age of people in a city.
4. **Statistic**: A statistic is a number that describes the sample. It is a known value, but it can change from sample to sample. example: average age of people in a sample.
7. **Variable**: A variable is any characteristic, number, or quantity that can be measured or counted. A variable may also be called a data item.
8. **Data**: Data is the information that you collect for the purpose of answering your research question. Data can be numbers, words, measurements, observations or even just descriptions of things.
9. **Mean**: The mean is the average of the numbers: a calculated "central" value of a set of numbers.
    Formulae: $\Large{\bar{x} = \frac{\sum_{i=1}^{n} x_i}{n}}$
10. **Median**: The median is the middle number in a sorted, ascending or descending, list of numbers and can be more descriptive of that data set than the average.
    Formulae: $\Large{M = \frac{n+1}{2}}$
11. **Mode**: The mode is the number that appears most often in a set of numbers.
    Formulae: The value that appears most frequently in a data set.
12. **Percentile**: A percentile is a measure used in statistics indicating the value below which a given percentage of observations in a group of observations fall.
    Formulae: $\Large{P = \frac{p}{100} \times (n+1)}$
13. **50-75-95-99 Percentile**: 50% of the data is below the 50th percentile, 75% of the data is below the 75th percentile, 95% of the data is below the 95th percentile, and 99% of the data is below the 99th percentile.
14. **Range**: The range is the difference between the largest and smallest numbers.
    Formulae: $\Large{Range = Max - Min}$
15. **Five Number Summary**: The five-number summary is a set of descriptive statistics that provides information about a dataset. It consists of the five most important sample percentiles: the minimum, the first quartile, the median, the third quartile, and the maximum.
    Formulae: $\Large{Min, Q1, M, Q3, Max}$
16. **simple random sampling vs stratified random sampling**: In simple random sampling, each member of the population has an equal chance of being selected as part of the sample. In stratified random sampling, the population is divided into subgroups, or strata, and random samples are taken from each stratum.
17. **Bias and chance error**: Bias is the systematic error that is introduced into sampling or testing by selecting or encouraging one outcome or answer over others. Chance error is the random error that is introduced into sampling or testing by choosing a sample or test group that is too small to be representative of the whole population.
18. **Different Bias**: Selection bias, Measurement bias, Recall bias, Reporting bias, Publication bias, etc.
     Examples: 
      - Selection bias: A sample of people who volunteer to participate in a study may not be representative of the population.
      - Measurement bias: A scale that is not properly calibrated.
      - Recall bias: People may not remember events accurately.
      - Reporting bias: People may not report events accurately.
      - Publication bias: Studies that show a significant effect are more likely to be published than studies that do not show a significant effect.
19. **Randomized Controlled Experiments**: Randomized controlled experiments are the gold standard for establishing cause-and-effect relationships. In a randomized controlled experiment, the researcher randomly assigns subjects to different conditions, and then manipulates the independent variable and measures the dependent variable.
20. **Dispersion/Variance**: Dispersion is a measure of how much the values in a dataset differ from the mean. It is a measure of the spread of the data.
    Formulae: $\Large{Dispersion = \frac{\sum_{i=1}^{n} (x_i - \bar{x})^2}{n}}$
21. **Standard Deviation**: The standard deviation is a measure of the amount of variation or dispersion of a set of values. A low standard deviation indicates that the values tend to be close to the mean of the set, while a high standard deviation indicates that the values are spread out over a wider range.
    Formulae: $\Large{\sigma = \sqrt{\frac{\sum_{i=1}^{n} (x_i - \bar{x})^2}{n}}}$



Probability
-----------

Probability is the likelihood of a specific outcome or event occurring out of all possible outcomes or events.
    Formulae: $\Large{P(A) = \frac{\text{Number of favorable outcomes}}{\text{Total number of possible outcomes}}}$

Simple Rules/Formulae of Probability:
-------------------------------------
  - P(A) + P(A') = 1
  - P(A and B) = P(A) * P(B) when A and B are independent
  - P(A or B) = P(A) + P(B) when A and B are mutually exclusive
  - P(A|B) = P(A and B) / P(B)


Bayes Theorem:
--------------
Bayes' theorem is a mathematical formula used to calculate conditional probability. It helps to revise existing predictions or theories given new or additional evidence.
    Formulae: $\Large{P(A|B) = \frac{P(B|A) \times P(A)}{P(B)}}$

    where P(B) can be written as P(B|A) * P(A) + P(B|A') * P(A')



Total Enumeration in probability:
---------------------------------
Total enumeration is a method of counting the number of ways that an event can occur. It is used to calculate the probability of an event.
    Example: Total number of ways to arrange 3 people in 3 chairs = 3! = 6



Warner's randomized response model
---------------------------------
Warner's randomized response model is a statistical model that allows respondents to answer sensitive questions without revealing their true answers. It is used to estimate the proportion of people who have a sensitive attribute, such as drug use or criminal behavior, without asking them directly about the sensitive attribute.

Here's how Warner's randomized response model could be implemented:

 - Randomization: Each respondent is given a coin or a similar randomization device (like a spinner or a random number generator) along with the survey.

 - Instructions: The respondents are instructed to flip the coin in private. If it lands heads, they are told to answer the question honestly. If it lands tails, they are instructed to provide a predetermined, randomized response, irrespective of the truth.

 - Answering the Sensitive Question: Let's say the sensitive question is "Have you ever cheated on an exam?" If the respondent gets heads, they answer truthfully (Yes or No). If they get tails, they follow a predetermined protocol that ensures their response remains confidential. For instance, they might be asked to respond with "Yes" regardless of their actual behavior.

 - Data Collection: Researchers collect the responses, not knowing which respondents were given heads or tails. This protects the privacy of individual respondents.

 - Analysis: Using statistical techniques, researchers can estimate the proportion of people who have the sensitive attribute (cheating on exams in this case) by taking into account the randomization process. By comparing the proportion of "Yes" responses from both the heads and tails groups, they can estimate the true proportion of individuals engaging in the sensitive behavior.

 The key to Warner's randomized response model is that even though respondents are asked to provide a randomized response when they get tails, the proportion of truthful answers can still be estimated using statistical methods. This provides a way to gather valuable information on sensitive topics while respecting respondent privacy and minimizing social desirability bias.



Normal Curve:
-------------
The normal curve is a bell-shaped curve that is symmetric about the mean. It is used to describe the distribution of a set of data. The normal curve is defined by two parameters: the mean and the standard deviation. The mean determines the location of the center of the curve, and the standard deviation determines the width of the curve.
Also known as Gaussian distribution, Normal distribution, or Bell curve.

Properties of Normal Curve:
  - The area under the curve is equal to 1.
  - The curve is symmetric about the mean.
  - The mean, median, and mode are all equal and located at the center of the curve.
  - The curve is bell-shaped, with the highest point at the mean.
  - The curve approaches but never touches the x-axis as it extends to infinity in both directions.
  - The curve is defined by two parameters: the mean and the standard deviation.
  - The standard deviation determines the width of the curve. A larger standard deviation results in a wider curve, and a smaller standard deviation results in a narrower curve.
  - The area under the curve between any two points on the x-axis represents the proportion of the data that falls between those two points.
  - The area under the curve to the left of any point on the x-axis represents the proportion of the data that falls to the left of that point.
  - The area under the curve to the right of any point on the x-axis represents the proportion of the data that falls to the right of that point.
  - The area under the curve between the mean minus one standard deviation and the mean plus one standard deviation represents approximately 68% of the data.
  - The area under the curve between the mean minus two standard deviations and the mean plus two standard deviations represents approximately 95% of the data.
  - The area under the curve between the mean minus three standard deviations and the mean plus three standard deviations represents approximately 99.7% of the data.

Standardizing the data
----------------------
Standardizing the data is the process of transforming the data so that it has a mean of 0 and a standard deviation of 1. This makes it easier to compare different sets of data.

Formulae:
  - $\Large{z = \frac{x - \bar{x}}{\sigma}}$
  - $\Large{x = \bar{x} + z \times \sigma}$

Properties of Standardized data:
  - The mean of the standardized data is 0.
  - The standard deviation of the standardized data is 1.
  - The standardized data has the same shape as the original data, but the units have been changed so that the mean is 0 and the standard deviation is 1.
  - The standardized data is easier to compare to other sets of data, because it has a mean of 0 and a standard deviation of 1.

Normal Approximation:
---------------------
The normal approximation is a method of approximating the distribution of a set of data with a normal curve. It is used to estimate the proportion of the data that falls within a certain range.
Formulae:
  - $\Large{P(a < x < b) = P(\frac{a - \bar{x}}{\sigma} < z < \frac{b - \bar{x}}{\sigma})}$
  - $\Large{P(a < x < b) = P(z_1 < z < z_2)}$
  - $\Large{z_1 = \frac{a - \bar{x}}{\sigma}}$
  - $\Large{z_2 = \frac{b - \bar{x}}{\sigma}}$

where: 
  - P(a < x < b) is the proportion of the data that falls between a and b.
  - P(z_1 < z < z_2) is the proportion of the data that falls between z_1 and z_2.
  - z_1 and z_2 are the standardized values of a and b.
  - a and b are the original values of the data.
  - $\bar{x}$ is the mean of the data.
  - $\sigma$ is the standard deviation of the data.

Square Root Law:
----------------
The square root law is a rule of thumb that states that the standard error of a sample statistic decreases by a factor of the square root of the sample size when the sample size is increased by a factor of the square of the original sample size.

  Formulae: $\Large{SE = \frac{\sigma}{\sqrt{n}}}$

where:
  - SE is the standard error.
  - $\sigma$ is the standard deviation of the population.
  - n is the number of samples.

  Standard error in sum increases by a factor of $\sqrt{n}$, when the sample size is increased by a factor of n.
